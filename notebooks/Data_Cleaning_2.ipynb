```python
# Update the functions to refine the extraction of OS type and version
def refined_extract_os_type(os):
    if 'Windows' in os:
        return 'Windows'
    elif 'Mac OS' in os or 'macOS' in os:
        return 'macOS'
    elif 'Chrome' in os or 'Google Chrome' in os:
        return 'Chrome OS'
    elif 'DOS' in os:
        return 'DOS'
    elif 'Jio OS' in os:
        return 'Jio OS'
    elif 'Prime OS' in os:
        return 'Prime OS'
    elif 'Linux' in os or 'Ubuntu' in os:
        return 'Linux'
    else:
        return 'Other'

def refined_extract_version(os):
    parts = os.replace('Google Chrome', '').replace('Home Basic', '').replace('Professional', '').strip().split(' ', 1)
    if len(parts) > 1:
        return parts[1].strip()
    return None

# Apply the refined functions to the dataset
df['OS Type'] = df['Operating System'].apply(refined_extract_os_type)
df['OS Version'] = df['Operating System'].apply(refined_extract_version)

# Check the result by displaying unique values for the new columns
refined_os_type_values = df['OS Type'].unique()
refined_os_version_values = df['OS Version'].unique()
refined_os_type_values, refined_os_version_values
```

```python
# Insert the new columns right after the original 'Operating System' column
# Find the index of the 'Operating System' column
os_index = df.columns.get_loc("Operating System")

# Insert the refined columns right after 'Operating System'
df.insert(os_index + 1, 'OS Type', df.pop('OS Type'))
df.insert(os_index + 2, 'OS Version', df.pop('OS Version'))

# Display the new arrangement of columns to confirm
df.columns
```

```python
# Step 1: Standardize 'Display Resolution' values
df['Display Resolution'] = df['Display Resolution'].str.replace('pixels', 'Pixels')

# Step 2: Extract Width and Height
# Inserting new columns after 'Display Resolution' column
df.insert(df.columns.get_loc('Display Resolution') + 1, 'Resolution Width', df['Display Resolution'].apply(lambda x: int(x.split(' x ')[0]) if pd.notnull(x) else None))
df.insert(df.columns.get_loc('Display Resolution') + 2, 'Resolution Height', df['Display Resolution'].apply(lambda x: int(x.split(' x ')[1].split()[0]) if pd.notnull(x) else None))

# Verify the changes
sample_rows = df[['Display Resolution', 'Resolution Width', 'Resolution Height']].head(10)

sample_rows
```

```python
# Step 3: Calculate Aspect Ratio
# Inserting the new column after 'Resolution Height' column
def calculate_aspect_ratio(width, height):
    if width and height:
        ratio = width / height
        return round(ratio, 2)
    return None

df.insert(df.columns.get_loc('Resolution Height') + 1, 'Aspect Ratio', df.apply(lambda row: calculate_aspect_ratio(row['Resolution Width'], row['Resolution Height']), axis=1))

# Verify the changes
sample_rows_with_aspect_ratio = df[['Display Resolution', 'Resolution Width', 'Resolution Height', 'Aspect Ratio']].head(10)

# Check null values count for all new columns
null_values_counts = df[['Resolution Width', 'Resolution Height', 'Aspect Ratio']].isnull().sum()

sample_rows_with_aspect_ratio, null_values_counts
```

```python
# Function to calculate PPI and round to 2 decimal places
def calculate_ppi(row):
    width = row['Resolution Width']
    height = row['Resolution Height']
    size = row['Display Size (Inches)']
    if width and height and size:
        diagonal_resolution = math.sqrt(width**2 + height**2)
        return round(diagonal_resolution / size, 2)
    return None

# Apply the function to calculate PPI and create a new column after 'Aspect Ratio'
df['PPI'] = df.apply(calculate_ppi, axis=1)

# Verify the changes
sample_rows_with_ppi_rounded = df[['Display Size (Inches)', 'Resolution Width', 'Resolution Height', 'Aspect Ratio', 'PPI']].head(10)

#sample_rows_with_ppi_rounded
```
